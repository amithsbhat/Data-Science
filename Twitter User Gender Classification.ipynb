```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.base import TransformerMixin
from sklearn.pipeline import Pipeline
```

```python
import string
from spacy.lang.en.stop_words import STOP_WORDS
from spacy.lang.en import English

# Create our list of punctuation marks
punctuations = string.punctuation

# Create our list of stopwords
nlp = spacy.load('en_core_web_sm')
stop_words = spacy.lang.en.stop_words.STOP_WORDS

# Load English tokenizer, tagger, parser, NER and word vectors
parser = English()

# Creating our tokenizer function
def spacy_tokenizer(sentence):
    # Creating our token object, which is used to create documents with linguistic annotations.
    tokens = parser(sentence)

    # Lemmatizing each token and converting each token into lowercase
    tokens = [ word.lemma_.lower().strip() if word.lemma_ != "-PRON-" else word.lower_ for word in tokens ]

    # Removing stop words
    tokens = [ word for word in tokens if word not in stop_words and word not in punctuations ]

    # return preprocessed list of tokens
    return tokens
```

```python
# Custom transformer using spaCy
class predictors(TransformerMixin):
    def transform(self, X, **transform_params):
        # Cleaning Text
        return [clean_text(text) for text in X]

    def fit(self, X, y=None, **fit_params):
        return self

    def get_params(self, deep=True):
        return {}

# Basic function to clean the text
def clean_text(text):
    # Removing spaces and converting text into lowercase
    return text.strip().lower()
```

```python
bow_vector = TfidfVectorizer(tokenizer = spacy_tokenizer, ngram_range=(1,1))
```

```python
from sklearn.model_selection import train_test_split

X = df['combined_data'].tolist()
y = df['gender'].tolist()
 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Logistic Regression Classifier
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression() 

#from sklearn.svm import LinearSVC
#classifier = LinearSVC()

#from xgboost import XGBClassifier
#classifier = XGBClassifier()


# Create pipeline using Bag of Words
pipe = Pipeline([("cleaner", predictors()),
                 ('vectorizer', bow_vector),
                 ('classifier', classifier)])

# model generation
pipe.fit(X_train,y_train)
```

```python
from sklearn import metrics
# Predicting with a test dataset
predicted = pipe.predict(X_train)

# Model Accuracy
print("Logistic Regression Train Accuracy:",metrics.accuracy_score(y_train, predicted))
print("Logistic Regression Train Precision:",metrics.precision_score(y_train, predicted, average='weighted'))
print("Logistic Regression Train Recall:",metrics.recall_score(y_train, predicted, average='weighted'))
```

```python
from sklearn import metrics
# Predicting with a test dataset
predicted = pipe.predict(X_test)

# Model Accuracy
print("Logistic Regression Test Accuracy:",metrics.accuracy_score(y_test, predicted))
print("Logistic Regression Test Precision:",metrics.precision_score(y_test, predicted, average='weighted'))
print("Logistic Regression Test Recall:",metrics.recall_score(y_test, predicted, average='weighted'))
```
