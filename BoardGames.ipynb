```python
df = pd.get_dummies(df, columns=['type'])
df.head()
```

```python

df['playingtime'] = df['playingtime'].fillna(df['playingtime'].mean())
df['minage'] = df['minage'].fillna(df['minage'].mean())

df['yearpublished'] = df['yearpublished'].fillna(df['yearpublished'].mean())
df['minplayers'] = df['minplayers'].fillna(df['minplayers'].mean())

df['maxplayers'] = df['maxplayers'].fillna(df['maxplayers'].mean())

```

```python
df.isnull().values.any()
```

```python
for col in df.columns:
    if df[col].isnull().values.any():
        print(col)
```

```python
print(df.columns)
```

```python
from sklearn.model_selection import train_test_split

feature_col_names = ['yearpublished', 'minplayers', 'maxplayers', 'playingtime', 'minage',
       'users_rated', 'bayes_average_rating', 'total_owners',
       'total_traders', 'total_wishers', 'average_weight', 'type_boardgame',
       'type_boardgameexpansion']
predicted_class_names = ['average_rating']

X = df[feature_col_names].values

Y = df[predicted_class_names].values
split_test_size = 0.3

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=split_test_size, random_state = 42)
```

```python

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train.ravel())


from sklearn.metrics import mean_squared_error
train_predictions = lr.predict(X_train)
test_predictions = lr.predict(X_test)

print("Train predictions accuracy for Linear Regression = {}".format(mean_squared_error(train_predictions, y_train)))
print("Test predictions accuracy for Linear Regression = {}".format(mean_squared_error(test_predictions, y_test)))
```

```python

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_train, y_train.ravel())


train_predictions = rf.predict(X_train)
test_predictions = rf.predict(X_test)

print("Train predictions accuracy for RandomForest Regression = {}".format(mean_squared_error(train_predictions, y_train)))
print("Test predictions accuracy for RandomForest Regression = {}".format(mean_squared_error(test_predictions, y_test)))
```

```python
import numpy as np
train_predictions_lr = lr.predict(X_train)
test_predictions_lr = lr.predict(X_test)

avg_train_pred = np.array((train_predictions_lr+train_predictions)/2.0)
avg_test_pred = np.array((test_predictions_lr+test_predictions)/2.0)


print("Train predictions accuracy for ensemble = {}".format(mean_squared_error(avg_train_pred, y_train)))
print("Test predictions accuracy for ensemble = {}".format(mean_squared_error(avg_test_pred, y_test)))
```
