```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

df = pd.read_csv(r"C:\\Contents\\Kaggle\\international-football-results-from-1872-to-2017\\results.csv")

df.head()
```

```python
df.date=df.date.apply(lambda x: x.split('-')[0])
```

```python
catcols = ['home_team','away_team','tournament', 'city', 'country','neutral']
df = pd.get_dummies(df, columns=catcols)
df.head()
```

```python
df.columns
```

```python
def getResult(h,a):
    h,a= int(h), int(a)
    if h==a:
        return 0
    elif h>a:
        return 1
    return 2

df['Result'] = df.apply(lambda row: getResult(row['home_score'], row['away_score']), axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
feature_col_names = [col for col in df.columns if col not in set(['home_score','away_score','Result'])]
predicted_class_names = ['Result']

X = df[feature_col_names].values

Y = df[predicted_class_names].values

split_test_size = 0.25

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=split_test_size, random_state = 42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
nb_model.fit(X_train, y_train.ravel())

nb_predict_train = nb_model.predict(X_train)
nb_predict_test = nb_model.predict(X_test)

from sklearn import metrics

print("NB Accuracy training : {0:.4f}".format(metrics.accuracy_score(y_train,nb_predict_train )))
print("NB Accuracy testing : {0:.4f}".format(metrics.accuracy_score(y_test,nb_predict_test )))
print('**********Confusion metrics**************')
print("{}".format(metrics.confusion_matrix(y_test, nb_predict_test)))
print('**********Classsification report**************')
print("{}".format(metrics.classification_report(y_test, nb_predict_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state = 42)
rf_model.fit(X_train, y_train.ravel())

rf_predict_train = rf_model.predict(X_train)
rf_predict_test = rf_model.predict(X_test)

from sklearn import metrics

print("RF Accuracy training : {0:.4f}".format(metrics.accuracy_score(y_train,rf_predict_train )))
print("RF Accuracy testing : {0:.4f}".format(metrics.accuracy_score(y_test,rf_predict_test )))
print('********** RF Confusion metrics**************')
print("{}".format(metrics.confusion_matrix(y_test, rf_predict_test)))
print('**********RF Classsification report**************')
print("{}".format(metrics.classification_report(y_test, rf_predict_test)))
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train.ravel())
lr_predict_train = lr_model.predict(X_train)
lr_predict_test = lr_model.predict(X_test)

from sklearn import metrics

print("LR Accuracy training : {0:.4f}".format(metrics.accuracy_score(y_train,lr_predict_train )))
print("LR Accuracy testing : {0:.4f}".format(metrics.accuracy_score(y_test,lr_predict_test )))
print('********** LR Confusion metrics**************')
print("{}".format(metrics.confusion_matrix(y_test, lr_predict_test)))
print('**********LR Classsification report**************')
print("{}".format(metrics.classification_report(y_test, lr_predict_test)))
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train.ravel())
lda_predict_train = lda.predict(X_train)
lda_predict_test = lda.predict(X_test)

from sklearn import metrics

print("LDA Accuracy training : {0:.4f}".format(metrics.accuracy_score(y_train,lda_predict_train )))
print("LDA Accuracy testing : {0:.4f}".format(metrics.accuracy_score(y_test,lda_predict_test )))
print('********** LDA Confusion metrics**************')
print("{}".format(metrics.confusion_matrix(y_test, lda_predict_test)))
print('**********LDA Classsification report**************')
print("{}".format(metrics.classification_report(y_test, lda_predict_test)))
```

```python
from xgboost.sklearn import XGBClassifier


xgb = XGBClassifier()
xgb.fit(X_train, y_train.ravel())
xgb_predict_train = xgb.predict(X_train)
xgb_predict_test = xgb.predict(X_test)

from sklearn import metrics

print("XGB Accuracy training : {0:.4f}".format(metrics.accuracy_score(y_train,xgb_predict_train )))
print("XGB Accuracy testing : {0:.4f}".format(metrics.accuracy_score(y_test,xgb_predict_test )))
print('********** XGB Confusion metrics**************')
print("{}".format(metrics.confusion_matrix(y_test, xgb_predict_test)))
print('**********XGB Classsification report**************')
print("{}".format(metrics.classification_report(y_test, xgb_predict_test)))
```

```python
import numpy as np
from scipy import stats

a = np.array([
  xgb_predict_test,
    lda_predict_test,
    lr_predict_test,
    rf_predict_test,
    nb_predict_test
])

ensemble_test = stats.mode(a)
```

```python
ensemble_test = np.array(ensemble_test[0])
```

```python
print("Ensemble Accuracy testing : {0:.4f}".format(metrics.accuracy_score(y_test,ensemble_test )))
print('********** Ensemble Confusion metrics**************')
print("{}".format(metrics.confusion_matrix(y_test, ensemble_test)))
print('**********Ensemble Classsification report**************')
print("{}".format(metrics.classification_report(y_test, ensemble_test)))
```
